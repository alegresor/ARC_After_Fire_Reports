```python
# method for making the graphs 
def gen_graph(df, graph_type, agg_var):
    if graph_type == 'bar_graph':
        total_injured_by_county = pd.pivot_table(df, values='People Injured', index='County',aggfunc=agg_var)
        total_unitsAffected_by_county = pd.pivot_table(df, values='Units Affected', index='County',aggfunc=agg_var)
        trace_injured = go.Bar(
            x = list(total_injured_by_county.index), 
            y = list(total_injured_by_county.values), 
            name = 'People Injured'
        )
        trace_unitsAffected = go.Bar(
            x = list(total_unitsAffected_by_county.index),
            y = list(total_unitsAffected_by_county.values),
            name = 'Units Affected'
        )
        my_data = [trace_injured,trace_unitsAffected]
        my_layout = go.Layout(barmode = 'group')
        fig = go.Figure(data = my_data, layout = my_layout)

        po.plot(fig,filename = 'testing2.html')
```

```python
# avg adults and children per incident in given df (pie chart). Allows volunteer to see the demographic of people living 
# in the area (more children or more adults, indication of family)


def gen_people_avg(df):
    adults_mean = df['Adults'].mean()
    children_mean = df["Children"].mean()
    
    categories = ['Adults','Children']
    
    vals = [adults_mean, children_mean]

    trace = go.Pie(labels = categories, values = vals)
    
    po.plot([trace])
    
```
