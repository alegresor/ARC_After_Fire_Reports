```python
# Incident Reports inital exploration
import pandas as pd
```

```python
df = pd.read_csv("../data/Fire_Stations.csv")
print(df.LOCATION.loc[0])
df.head()
```

```python
# Split off lat,lng categories
def splitLocation(location):
    address,cityStateZip,latLng = location.split('\n')
    lat,lng = latLng.strip('()').split(', ')
    return float(lat),float(lng)
df['lat'],df['lng'] = zip(*df['LOCATION'].map(splitLocation))
df = df.drop(['LOCATION'],axis='columns')
df.head()
```

```python
df.to_csv('../data/Fire_Stations_Clean.csv')
```
